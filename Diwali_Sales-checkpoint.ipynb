```python
# Analysis on Basis of Gender
```

```python
axe=sns.countplot(x='Gender',data=df)
```

```python
axe=sns.countplot(x='Gender',data=df)

for bars in axe.containers:
    axe.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales)
```

```python
# Above Graphs show most of the buyers are females and purchasing power of females is more
```

```python
'''Analysis on basis of Age'''
```

```python
df.columns
```

```python
axe = sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in axe.containers:
    axe.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
# Above shows most of the buyers are 26-35 female buyers
```

```python
# Analysis on basis of State
```

```python
# Total number of orders from top 5 States
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
#Total amount spend on shopping from top 5 states
```

```python
sales_amt = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_amt,x='State',y='Amount')
```

```python
# Graph shows most of the orders are from Uttar Pradesh , Maharashtra & Karnataka state Similarly the amount spend also.
```

```python
# Analysis based upon martial_status
```

```python
df.columns
```

```python
axe = sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(6,5)})

for bars in axe.containers:
    axe.bar_label(bars)
```

```python
sales_marital = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})

sns.barplot(data=sales_marital, x='Marital_Status',y='Amount',hue='Gender')
```

```python
# Most of the shopping is done by married(females)that also have high purchasing power
```

```python
# Ansalysis on basis of occupation
```

```python
axe = sns.countplot(data=df,x='Occupation')

sns.set(rc={'figure.figsize':(20,10)})
```

```python
sns.set(rc={'figure.figsize':(20,5)})
axe = sns.countplot(data=df,x='Occupation')
```

```python
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=sales_occ,x='Occupation',y='Amount')
```

```python
# Above plot shows people engaged within IT sector as occupation have highest purchasing power
```

```python
# Analysis based upon product_category
```

```python
df.columns
```

```python
axe = sns.countplot(data = df,x='Product_Category')
sns.set(rc={'figure.figsize':(40,20)})

for bars in axe.containers:
    axe.bar_label(bars)
```

```python
sales_prod = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(40,20)})
sns.barplot(data=sales_prod,x='Product_Category',y='Amount')
```

```python
# From above both graphs conclusion obtained that most of the orders are done for clothing then food but 
# Most of the amount spent is for food and then clothing
```

```python
# Analysis of top selling products on basis of product_id
```

```python
df.columns
```

```python
sales_prod_id = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_prod_id,x='Product_ID',y='Orders')
```

**Conclusion**

*Most of the orders are placed by married womens of age between 26-35 from Uttar Pradesh, Maharastra & Karnataka State working in IT Sector , Healthcare & Aviation buy Products from Food, Clothing & Electronics Category*
